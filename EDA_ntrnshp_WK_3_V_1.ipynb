```python
import matplotlib.pyplot as plt
```

# Line Graph

```python
# Given data points
x_values = [1, 2, 3]
y_values = [2, 4, 1]

# Plotting the line
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph')

# Display the graph
plt.show()
```

# Multi-Line Graph

```python
# Data sets
data_set_21 = [[10, 40], [15, 25], [20, 10], [25, 20], [30, 30]]
data_set_22 = [[10, 20], [20, 40], [25, 20], [30, 10]]
data_set_23 = [[10, 30], [20, 25], [30, 10]]

# Plotting the lines
plt.plot(*zip(*data_set_21), color='red', linestyle='-', linewidth=2, label='Data Set 1')
plt.plot(*zip(*data_set_22), color='blue', linestyle='--', linewidth=1.5, label='Data Set 2')
plt.plot(*zip(*data_set_23), color='green', linestyle=':', linewidth=2.5, label='Data Set 3')

# Adding labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines with Legend')
plt.legend()

# Display the graph inline
plt.show()
```

# Multi-DataSet Graph

```python
# Data sets
data_set_31 = [[1, 2], [4, 6], [5, 3], [6, 6], [7, 3]]
data_set_32 = [[1, 3], [3, 4], [4, 5],[5,6], [7, 7]]

# Plotting the lines with different markers
plt.plot(*zip(*data_set_31), marker='s', linestyle='-', color='blue', label='Data Set 1')
plt.plot(*zip(*data_set_32), marker='^', linestyle='-', color='green', label='Data Set 2')

# Adding labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Different Markers')
plt.legend()

# Display the graph inline in Jupyter
plt.show()
```

# BarPlot From Data

```python
# Programming languages and their popularity
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
# Plotting the horizontal bar chart
plt.barh(languages, popularity, color='green')

# Adding a grid
#plt.grid(True)
plt.grid(axis='both', linestyle='--', alpha=0.7, color='red', linewidth=1)

# Adding labels and title
plt.xlabel('Popularity (%)')
plt.title('Popularity of Programming Languages')

# Display the chart
plt.show()
```

## BarPlot Using SeaBorn 

```python
import pandas as pd
import seaborn as sns
```

```python
# Create a DataFrame using the sample data
df = pd.DataFrame({'Languages': languages, 'Popularity': popularity})

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.barplot(x='Popularity', y='Languages', data=df, palette='viridis')

# Manually add grid lines on both axes
plt.grid(axis='both', linestyle='--', alpha=1, color='red', linewidth=1)

# Adding labels and title
plt.xlabel('Popularity (%)')
plt.ylabel('Programming Languages')
plt.title('Popularity of Programming Languages')

# Display the chart
plt.show()
```

# Multi-BarPlot From DataFrame

```python
# Sample data
data_samp5 = {
    'Category': [2, 4, 6, 8, 10],
    'a': [4, 2, 4, 2, 2],
    'b': [8, 3, 7, 6, 4],
    'c': [5, 4, 4, 4, 3],
    'd': [7, 2, 7, 8, 3],
    'e': [6, 6, 8, 6, 2]
}

# Create a DataFrame
df_5 = pd.DataFrame(data_samp5)

# Melt the DataFrame to long format
df_long = pd.melt(df_5, id_vars='Category', var_name='Value Group', value_name='Values')

# Set Seaborn style
sns.set(style="whitegrid")

# Create a bar plot with multiple plots
plt.figure(figsize=(12, 8))
sns.barplot(x='Category', y='Values', hue='Value Group', data=df_long, palette='deep')

# Adding labels and title
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Seaborn Bar Plot with Multiple Plots')

# Display the legend
plt.legend(title='Value Group')

# Display the chart
plt.show()
```
